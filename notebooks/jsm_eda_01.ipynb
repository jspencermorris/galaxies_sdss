```python
import pandas as pd
```

```python
# load tabular data
# NOTE: there is one more dataset (original galaxy zoo 1) that also includes overlapping galaxy classifications, with labels spiral, elliptital, irregular...
filename_mapping = pd.read_csv('../data/raw/gz2/gz2_filename_mapping.csv', header=0) # galaxies - includes mapping definition between image filenames and (dr7)objid
hart16 = pd.read_csv('../data/raw/gz2/gz2_hart16.csv', header=0) # galaxies - includes detailed (sub)classification labels using dr7objid
zooVotes = pd.read_csv('../data/raw/skyserver_sdss/skyserver_sdss_dr16_zooVotes.csv', header=0) # galaxies - includes mapping definition between dr7objid and (dr8)objid
PhotoObj = pd.read_csv('../data/raw/skyserver_sdss/skyserver_sdss_dr16_PhotoObj.csv', header=0) # general - includes (super)classification labels and spectral intensities using (dr8)objid
specObj = pd.read_csv('../data/raw/skyserver_sdss/skyserver_sdss_dr16_specObj.csv', header=0) # general - includes (super)classification labels and photomask intensities using (dr8)bestobjid
```

```python
# rename identifier columns for improved clarity
filename_mapping.rename(columns={'objid':'dr7objid'}, inplace=True)
PhotoObj.rename(columns={'objID':'dr8objid', 'type':'sdss_class_number'}, inplace=True)
specObj.rename(columns={'bestObjID':'dr8objid', 'class':'sdss_class_string'}, inplace=True)
zooVotes.rename(columns={'objid':'dr8objid'}, inplace=True)
```

```python
# display data heads
display(filename_mapping.head(3))
display(hart16.head(3)) # main features: [];  main labels: [gz2_class]
display(zooVotes.head(3)) # main features: [];  main labels: [p_el, p_cw, p_acw, p_edge, p_dk, p_mg, p_cs]
display(PhotoObj.head(3)) # main features: [clean, u, g, r, i, z];  main labels: [sdss_class_number]
display(specObj.head(3)) # main features: [z, spectro...(all), elodie...(all)];  main labels: [sdss_class_string, subClass, class_noqso]
```

```python
# display dataframe shapes
display(filename_mapping.shape)
display(hart16.shape)
display(zooVotes.shape)
display(PhotoObj.shape)
display(specObj.shape)
```

```python
# determine number of unique labels
print(len(hart16['gz2_class'].unique())) # each class is a galaxy morphology subclass
print(len(PhotoObj['sdss_class_number'].unique())) # 3=galaxy, 6=star
print(len(specObj['sdss_class_string'].unique())) # galaxy, star, qso
```

```python
# evaluate the number of overlapping records in different tables

# first and second tables
merged_df = pd.merge(filename_mapping, hart16, left_on='dr7objid', right_on='dr7objid', how='inner')
print(f"Matching values: {len(merged_df['dr7objid'].tolist())}")
print(f"Percentage of matching values in df1: {(len(merged_df) / len(filename_mapping)) * 100}%")
print(f"Percentage of matching values in df2: {(len(merged_df) / len(hart16)) * 100}%")

# first and third tables
merged_df = pd.merge(filename_mapping, zooVotes, left_on='dr7objid', right_on='dr7objid', how='inner')
print(f"Matching values: {len(merged_df['dr7objid'].tolist())}")
print(f"Percentage of matching values in df1: {(len(merged_df) / len(filename_mapping)) * 100}%")
print(f"Percentage of matching values in df2: {(len(merged_df) / len(zooVotes)) * 100}%")

# second and third tables
merged_df = pd.merge(hart16, zooVotes, left_on='dr7objid', right_on='dr7objid', how='inner')
print(f"Matching values: {len(merged_df['dr7objid'].tolist())}")
print(f"Percentage of matching values in df1: {(len(merged_df) / len(hart16)) * 100}%")
print(f"Percentage of matching values in df2: {(len(merged_df) / len(zooVotes)) * 100}%")

# first and fourth tables

# second and fourth tables

# third and fourth tables
merged_df = pd.merge(zooVotes, PhotoObj['dr8objid'], left_on='dr8objid', right_on='dr8objid', how='inner')
print(f"Matching values: {len(merged_df['dr8objid'].tolist())}")
print(f"Percentage of matching values in df1: {(len(merged_df) / len(zooVotes)) * 100}%")
print(f"Percentage of matching values in df2: {(len(merged_df) / len(PhotoObj)) * 100}%")

# fourth and fifth tables
merged_df = pd.merge(PhotoObj['dr8objid'], specObj['dr8objid'], left_on='dr8objid', right_on='dr8objid', how='inner')
print(f"Matching values: {len(merged_df['dr8objid'].tolist())}")
print(f"Percentage of matching values in df1: {(len(merged_df) / len(PhotoObj)) * 100}%")
print(f"Percentage of matching values in df2: {(len(merged_df) / len(specObj)) * 100}%")
```
